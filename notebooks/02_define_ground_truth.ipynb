# 02 — Define Hidden Ground Truth

Create a synthetic "hidden truth" by perturbing specific kinetic parameters from their nominal (textbook) values. This simulates a biological system that deviates from literature.

**Output:** `ground_truth/true_parts.pkl`

```python
import os
# Navigate to repo root so all relative paths resolve correctly.
if os.path.basename(os.getcwd()) in ('notebooks', 'examples'):
    os.chdir('..')
```

```python
import pickle
import numpy as np

# Load nominal (textbook) parameters
with open(os.path.join("gcad", "data", "parts.pkl"), 'rb') as f:
    nominal_parts = pickle.load(f)

print(f"Loaded {len(nominal_parts)} nominal parts.")
```

## Define Mutations

Edit the cells below to change which parameters are perturbed and by how much.
All un-listed parts remain at their nominal (textbook) values.

```python
true_parts = {k: v.copy() if hasattr(v, 'copy') else v for k, v in nominal_parts.items()}

# Mutation 1: Z6 (Activator) — Kd 25% stronger, ymax 25% stronger
old_z6 = nominal_parts['Z6'].copy()
true_parts['Z6'] = np.array([0.015, 72.5, old_z6[2]])
print(f"Z6 Nominal: {old_z6}")
print(f"Z6 Truth:   {true_parts['Z6']}")

# Mutation 2: I13 (Inhibitor) — Kd 25% stronger
old_i13 = nominal_parts['I13'].copy()
true_parts['I13'] = np.array([0.009])
print(f"\nI13 Nominal: {old_i13}")
print(f"I13 Truth:   {true_parts['I13']}")
```

## Save

```python
os.makedirs(os.path.join("data", "ground_truth"), exist_ok=True)
with open(os.path.join("data", "ground_truth", "true_parts.pkl"), 'wb') as f:
    pickle.dump(true_parts, f)
print("Saved true_parts.pkl to ground_truth/")
```
